三種狀況:
1.	刪除節點沒有子節點的狀況,只要刪除目標節點就好
2.	刪除節點只有一個子節點時,先刪除目標節點,再被刪除節點的子節點移動到被刪除節點的位置
3.	刪除節點有兩個子節點時，先從右子樹中,找到最小值,因為右子樹中最小值可以剛好大於被刪除節點，且可以滿足bst的條件。

如果我今天要刪除6：
先確認有沒有該刪除節點:如果搜尋不到就回傳None代表沒找到

首先我要先找到6,將6依序從root開始比對,如果比root小就往左子樹走,確認有沒有左子樹,接著,與左子樹根節點比較,比左子樹根節點大往左子樹跟節點的右子樹走,確認有沒有左子樹根節點的右子樹根節點,.接著與左子樹根節點的右子樹根節點比較,比左子樹根節點的右子樹跟節點大,往他的右子樹走，比對一下
如果相等,就代表找到
確認有該刪除節點之後在執行判斷
第二步,確認該刪除節點狀態，
寫判斷式判斷為下列何種狀況：
1.如果該刪除節點,檢查確認left & right是null,就將該刪除節點的根節點的pointer改成null 。
2.如果該刪除節點,檢查left & right , 如果有其中一個, 則將該刪除節點的根節點指向改刪除節點的子節點。
3.如果該刪除節點,檢查left & right,  如果兩個都有滿足，將該刪除節點右子樹
中最小值替代為該刪除節點。
